```python
import pandas as pd
import math
import os
import shutil
import zipfile
import gzip
import numpy as np
```

```python
csv_inceptionv3_with_outliers = './predicts/pred_fine_tuning_inceptionv3_with_outliers_final.csv'
csv_xception_with_outliers = './predicts/pred_fine_tuning_xception_with_outliers_final.csv'
csv_inceptionresnetv2_with_outliers = './predicts/pred_fine_tuning_inceptionresnetv2_with_outliers_final.csv'
csv_inceptionv3_no_outliers = './predicts/pred_fine_tuning_inceptionv3_no_outliers_final.csv'
csv_xception_no_outliers = './predicts/pred_fine_tuning_xception_no_outliers_final.csv'
csv_inceptionresnetv2_no_outliers = './predicts/pred_fine_tuning_inceptionresnetv2_no_outliers_final.csv'
csv_sample = './predicts/sample_submission.csv'

csv_merged_models_with_outliers = './predicts/pred_merged_models_with_outliers_final.csv'
csv_merged_models_no_outliers = './predicts/pred_merged_models_no_outliers_final.csv'

df1= pd.read_csv(csv_inceptionv3_with_outliers)
df2= pd.read_csv(csv_xception_with_outliers)
df3= pd.read_csv(csv_inceptionresnetv2_with_outliers)
df4= pd.read_csv(csv_sample)

probilities= (np.array(df1['label'], dtype=np.float16)+np.array(df2['label'],dtype=np.float16)+np.array(df3['label'],dtype=np.float16))/3
probilities = probilities.clip(min=0.005, max=0.995)
print(probilities)
print(probilities.shape)
df4['id'] =  df4[['id']].apply(np.int32)

for i, p in enumerate(probilities):
    df4.at[i,'label'] = p
df4.to_csv(csv_merged_models_with_outliers, index = None)
df4['id'] =  df4[['id']].apply(np.int32)
print(df4)

print("finished")
```

```python
df1= pd.read_csv(csv_inceptionv3_no_outliers)
df2= pd.read_csv(csv_xception_no_outliers)
df3= pd.read_csv(csv_inceptionresnetv2_no_outliers)
df4= pd.read_csv(csv_sample)
df4['label']= np.mean(df1['label']+df2['label']+df3['label'], dtype=np.float16)
df4.to_csv(csv_merged_models_no_outliers)
print("finished")
```
